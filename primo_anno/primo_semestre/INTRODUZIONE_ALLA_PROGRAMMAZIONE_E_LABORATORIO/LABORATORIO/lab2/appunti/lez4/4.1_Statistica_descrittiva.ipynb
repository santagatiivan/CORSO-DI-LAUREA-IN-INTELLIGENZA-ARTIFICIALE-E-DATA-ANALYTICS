# Concetti base di probabilitÃ 

Cos'Ã¨ la **probabilitÃ  di un evento**?

Ci sono avvenimenti che accadono con certezza, mentre altri sicuramente non possono mai verificarsi. Per esempio, se una scatola contiene soltanto palline nere, estraendone una a caso siamo sicuri che Ã¨ nera, mentre Ã¨ impossibile estrarre una pallina bianca.
Chiamiamo gli avvenimenti del primo tipo **eventi certi** e quelli del secondo tipo **eventi impossibili**.
Ci sono anche eventi che possono accadere, ma senza certezza. Se la scatola contiene sia palline bianche sia palline nere, lâestrazione di una pallina bianca Ã¨ un evento possibile ma non certo, cosÃ¬ come lâestrazione di
una pallina nera. In altre parole, non possiamo prevedere il colore della
pallina estratta, perchÃ© lâestrazione Ã¨ **casuale**.

Come la rappresentiamo?
Chiamiamo **casi possibili** tutti i risultati che possono verificarsi.
Chiamiamo **casi favorevoli** quelli in cui si verifica lâevento che mi interessa.

Il quoziente $\frac{\text{numero casi favorevoli}}{\text{numero casi possibili}}$ viene chiamato **probabilitÃ ** di quellâevento.
E' quindi un valore tra 0 (evento impossibile) e 1 (evento certo).

Se moltiplichiamo questo valore per 100 otteniamo la probabilitÃ  espressa come percentuale (%).

La statistica si pone lo scopo di stimare i parametri che descrivono la dinamica aleatoria a partire da dati, ossia da osservazioni ripetute del sistema reale, che il modello stocastico si propone di approssimare.

# Distribuzioni

I dati raccolti possono essere classificati attraverso distribuzioni.
La distribuzione Ã¨ una rappresentazione del modo in cui le diverse modalitÃ /valori di un carattere si distribuiscono.

Le distribuzioni possono essere discrete o continue.

## Distribuzioni (di probabilitÃ ) per valori discreti (finiti)

Definite su un insieme a valori discreti.

Lo spazio degli stati S, Ã¨ composto da un numero discreto di modalitÃ /valori (e.g. i diversi colori degli occhi).

Ad ogni elemento Ã¨ associata una probabilitÃ  di essere pescato, $p(x), x \in S$. Queste probabilitÃ  devono sommare a $1$.

Esempio di una sacca con 20 palline di colori diversi (1 nera, 4 blu, 5 verdi, 10 marroni).

$p(\text{nera}) = \frac{1}{20}, \quad p(\text{blu}) = \frac{4}{20}, \quad p(\text{verde}) = \frac{5}{20}, \quad p(\text{marrone}) = \frac{10}{20}$

$p(\text{nera}) = 0.05, \quad p(\text{blu}) = 0.2, \quad p(\text{verde}) = 0.25, \quad p(\text{marrone}) = 0.5$


1/20+4/20+5/20+10/20=1

### Moda, media, mediana

*   La **Moda** Ã¨ definita come il valore che ha la frequenza piÃ¹ alta

La media e la mediana si possono calcolare solo su  distribuzioni di un carattere quantitativo oppure qualitativo ordinabile.

*   La **Media (aritmetica)** Ã¨ quel valore che corrisponde alla somma di tutti i valori diviso il numero dei valori stessi.
*   La **Mediana** Ã¨ quel valore/modalitÃ  assunto dalle unitÃ  statistiche che si trovano nel mezzo della distribuzione

Come si calcola la mediana:

*   Si ordinano gli n dati in modo crescente
*   se il numero di dati Ã¨ dispari la mediana corrisponde al valore centrale, ovvero al valore che occupa la posizione (n+1)/2
*   se il numero n di dati Ã¨ pari, la mediana Ã¨ stimata utilizzando i due valori che occupano le posizioni n/2 e n/2 +1 (generalmente si sceglie la loro media aritmetica se il carattere Ã¨ quantitativo).



````python
colori = np.array(['18', '19', '20', '21', '22'])
vettore_probabilita = np.array([1/20,9/20,6/20,3/20,1/20])
vettore_valori=np.array([18,19,19,19,19,19,19,19,19,19,20,20,20,20,20,20,21,21,21,22])
n = len(vettore_valori)
media = np.sum(vettore_valori)/n
if n%2==0:  # se n Ã¨ pari
  mediana = (vettore_valori[n//2-1] + vettore_valori[n//2])/2.  # considerando l'indice che parte da zero quindi devo "slittare" di uno indietro
else:
  mediana = vettore_valori[(n-1)//2] # (n+1)/2 -1 , considerando l'indice che parte da zero quindi devo "slittare di uno indietro"
print('media:', media)
print('mediana:', mediana)

#esistono i metodi in numpy
print('media:', np.mean(vettore_valori))
print('mediana:', np.median(vettore_valori))
````

## Distribuzioni (di probabilitÃ ) per valori continui

Definite su un insieme continuo. Ad esempio un intervallo dei numeri reali.

### Distribuzione Uniforme
La probabilitÃ  ha un valore costante in un dato intervallo [a,b] e zero fuori dall'intervallo. Si indica di solito con U(a,b) dove a e b sono gli estremi dell'intervallo. Il valore costante Ã¨ pari a 1/(b-a).

````python
X = st.uniform(0,1)
X2 = st.uniform(2,2)
fig, ax = plt.subplots()
xs = np.linspace(-0.1, 4.1, 200)
ax.plot(xs, X.pdf(xs),label='U[0,1]')
ax.plot(xs, X2.pdf(xs),label='U[2,4]')
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')
plt.title('distribuzione uniforme continua')
ax.legend()
````



### Distribuzione Normale (o di Gauss)
Serve per descrivere variabili casuali a valori reali che tendono a concentrarsi attorno a un singolo valor medio (teorema del limite centrale).

Ha una forma come di una campana rovesciata dove la punta della campana corrisponde proprio al valore medio.

La curva Ã¨ simmetrica rispetto al valore medio

Moltissime modalitÃ  attorno a voi hanno una distribuzione normale. Ad esempio l'altezza ed il peso delle persone, od i voti scolastici.

La distribuzione normale dipende da due parametri, la media $\mu$ e la varianza $\sigma^2$, ed Ã¨ indicata tradizionalmente con: $N(\mu, \sigma^2)$,

La varianza Ã¨ una misura della dispersione dei valori di una variabile statistica rispetto alla sua media, cioÃ¨ Ã¨ la misura di quanto i valori si discostano (quadraticamente) dalla media aritmetica.

La varianza di un insieme di dati $(x_1, x_2, \dots, x_n)$ Ã¨ definita come:
$
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
$

dove $n$ Ã¨ il numero totale di elementi, $x_i$ Ã¨ il valore dell'i-esima osservazione, $\mu$ Ã¨ la media dei valori, cioÃ¨ $\mu = \frac{1}{n} \sum_{i=1}^{n} x_i $.

La sua radice quadrata, $\sigma$, si chiama deviazione standard.

````python
mu = 175.0
sigma = 10.0
X = st.norm(loc=mu, scale=sigma)

fig, ax = plt.subplots()
xs = np.linspace(mu - 6.0 * sigma, mu + 6.0 * sigma, 100)
ax.plot(xs, X.pdf(xs))
ax.axvline(np.mean(xs), color='r', linestyle='--', label="Mean")
ax.axvline(np.median(xs), color='g', linestyle='-', label="Median")
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')
plt.xlim([100, 250])
ax.legend()
````

````python
def plot_metrics(campione, titolo, nret=10):
  df = pd.DataFrame(campione, columns = ['samples'])

  f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
  mean=df['samples'].mean()
  median=df['samples'].median()


  plt.title(titolo)
  sns.boxplot(data=df, x="samples", ax=ax_box)
  ax_box.axvline(mean, color='r', linestyle='--')
  ax_box.axvline(median, color='g', linestyle='-')

  sns.histplot(data=df, x="samples", ax=ax_hist, bins=nret)
  ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
  ax_hist.axvline(median, color='g', linestyle='-', label="Median")

  ax_hist.legend()

  ax_box.set(xlabel='')

  plt.show()
  plt.close()
````

**Distribuzione Uniforme**

````python
# 10 campioni
print("Media: " + str(np.mean(campione_uniforme_10)))
print("Mediana: " + str(np.median(campione_uniforme_10)))

plot_metrics(campione_uniforme_10,'UNIFORME: 10 campioni')

# 100 campioni
print("\nMedia: " + str(np.mean(campione_uniforme_100)))
print("Mediana: " + str(np.median(campione_uniforme_100)))
plot_metrics(campione_uniforme_100,'UNIFORME: 100 campioni')

# 10k campioni
print("\nMedia: " + str(np.mean(campione_uniforme_10K)))
print("Mediana: " + str(np.median(campione_uniforme_10K)))
plot_metrics(campione_uniforme_10K, 'UNIFORME: 10K campioni')
````

### Distribuzione asimmetrica

## Legge dei grandi numeri
La legge dei grandi numeri Ã¨ un teorema fondamentale della teoria della probabilitÃ  che indica che se ripetiamo molte volte (tendendo all'infinito) lo stesso esperimento, la frequenza di un certo evento tende ad essere costante.

Vale a dire, **la legge dei grandi numeri indica che se lo stesso test viene eseguito ripetutamente (ad esempio, lancio di una moneta), la frequenza con cui si ripeterÃ  un determinato evento (testa/croce) si avvicinerÃ  a una costante. Questa a sua volta sarÃ  la probabilitÃ  che questo evento si verifichi.**
